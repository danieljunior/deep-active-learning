```python
import pandas as pd
```

```python
data = pd.read_csv('../data/all_runs.csv', sep=";")
nsp_models_data = data[data.tag == 'only_nsp_v0']
baselines_data = data[data.tag == 'only_baselines_v0']
```

```python
set(nsp_models_data.strategy.values)
```

```python
set(nsp_models_data.columns.values)
```

```python
pd.options.display.max_rows = None
nsp_models_data[['model','strategy','dataset']].drop_duplicates()
```

```python
pd.options.display.max_rows = 10

local_stj_runs = nsp_models_data[nsp_models_data.dataset.isin(['LOCAL_STJ'])]
rs_local_stj = local_stj_runs[local_stj_runs.strategy == 'RandomSampling']
rs_local_stj
```

```python
data1 = rs_local_stj
# data1 = data1[data1['round'] == 5.0]
data1 = data1[data1.train_batch_size == 32]
data1 = data1[data1.samples == 5120.0]
data1 = data1[data1.n_query == 64]
data1 = data1[data1.n_init_labeled == 128.0]
data1
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
sns.set(style="whitegrid")

#Graph is created and stored in the variable 'graph'
g = sns.barplot(x="total_labeled_data",y="test_accuracy",hue='model',data=data1)
g.set_title("Strategy = {} - Total Samples = {} - Initial Pool = {} - Query = {} - Train batch = {}".format(
'Random Sampling', 5120, 128, 64, 32))
g.set_xticklabels(g.get_xticklabels(), rotation=90)

for idx, row in baselines_data[(baselines_data.dataset == 'LOCAL_STJ') &
              (baselines_data.samples == 5120.0)].iterrows():
    g.axhline(row['test_accuracy'], label=row['model'] +' -> '+ row['strategy'],
             ls='--', color=list(mcolors.CSS4_COLORS.keys())[idx])
        
#Drawing a horizontal line at point 1.25
#The plot is shown
plt.legend()
sns.move_legend(g, "upper left", bbox_to_anchor=(1, 1))
plt.show()
```
